Imagine you buy a powerful car.

## 🔹 What is Anaconda Navigator?

Anaconda Navigator is a **graphical user interface (GUI)** that comes with:

* Anaconda

It allows you to:

* Manage environments visually
* Install packages via search
* Launch applications
* Switch between environments

It works on top of Conda.

Conda = engine
Navigator = dashboard

# ✅ Answers – 2.5 Anaconda Navigator Tour

---

## 🟢 Basic Exercise 1

**Where to create a new environment?**

* Go to the **Environments** tab
* Click **Create**
* Enter environment name
* Select Python version
* Click **Create**

That creates an isolated environment visually.

---

## 🟢 Basic Exercise 2

**Installed vs Available Packages**

* **Installed:** Packages already present in the selected environment.
* **Available:** All packages that can be installed from Conda repositories.

Installed = currently active tools

Available = optional tools you can add

---

## 🟡 Intermediate Exercise 1

**Fixing wrong Pandas version using Navigator (GUI):**

1. Go to **Environments tab**
2. Select the correct environment
3. Search for “pandas”
4. Change filter from *Installed* to *All*
5. Choose correct version from dropdown
6. Click **Apply**

Navigator will downgrade/upgrade safely.

---

## 🟡 Intermediate Exercise 2

**Why Navigator may slow workflow in industry:**

1. GUI is slower than command-line
2. Not scriptable or automatable
3. Difficult to reproduce steps exactly
4. Cannot integrate with CI/CD pipelines

Professionals prefer terminal for speed and automation.

---

## 🔴 Advanced Exercise 1

### Bootcamp Strategy (50 Students)

**Stage 1 – Beginner Level**

* Start with Navigator
* Reduce terminal fear
* Focus on concepts

**Stage 2 – Intermediate**

* Introduce terminal commands
* Show equivalent Conda CLI commands

**Stage 3 – Advanced**

* Fully transition to terminal
* Teach environment exporting
* Teach automation scripts

Progression:
GUI → Mixed → CLI mastery

This builds confidence gradually.

---

## 🔴 Advanced Exercise 2

**Why Navigator is unsuitable for CI/CD systems**

CI/CD requires:

* Automation
* Script-based installations
* Reproducibility
* Version locking
* Infrastructure-as-code

Navigator:

* Requires manual clicking
* Cannot be version-controlled
* Not scalable
* Not usable in headless cloud servers

In production pipelines:

* `environment.yml`
* `conda install`
* Dockerfiles
* Automated scripts

are standard.
