### 🔹 What is Jupyter Notebook?

Jupyter Notebook is an interactive coding environment where you can:

* Write code
* Write text explanations
* Run code cell by cell
* Visualize graphs
* Display tables
* Mix documentation + execution

Everything in one place.

---

### 🔹 What is JupyterLab?

JupyterLab is the **modern upgraded version** of Jupyter Notebook.

Think of it like:

Notebook = Classic classroom
JupyterLab = Full university campus

JupyterLab includes:

* Multiple notebooks in tabs
* File browser
* Terminal access
* Text editor
* Better layout management

---

## 🔎 Key Concept: Cells

Jupyter works using **cells**.

There are 2 main types:

### 🟢 Code Cell

Executes Python code.

Shortcut:

* `Shift + Enter` → Run cell

---

### 🔵 Markdown Cell

Write formatted text:

* Headings
* Bullet points
* Explanations
* LaTeX math

Example:

```markdown
# My Project
This is churn prediction analysis.
```

# ✅ Answers – 2.6 JupyterLab & Jupyter Notebook

---

## 🟢 Basic Exercise 1

**Difference between Code Cell and Markdown Cell:**

* **Code Cell:** Executes Python code and produces output.
* **Markdown Cell:** Used for writing formatted text, explanations, headings, and documentation.

Code = execution
Markdown = explanation

---

## 🟢 Basic Exercise 2

**Why running cell-by-cell is useful:**

1. Debug errors step-by-step
2. Test small parts of code
3. Analyze intermediate results
4. Experiment safely without running full script

This supports exploratory analysis.

---

## 🟡 Intermediate Exercise 1

**Why errors happened?**

Because Jupyter executes cells in the order you run them, not top-to-bottom automatically.
Some variables were not defined or were overwritten.

**Fix:**

1. Restart kernel
2. Run all cells from top to bottom
3. Maintain logical execution order

Menu → Restart & Run All

---

## 🟡 Intermediate Exercise 2

**Proper Notebook Structure Using Markdown:**

```
# Project Title: Customer Churn Prediction

## 1. Import Libraries

## 2. Load Dataset

## 3. Data Cleaning

## 4. Exploratory Data Analysis

## 5. Feature Engineering

## 6. Model Training

## 7. Model Evaluation

## 8. Conclusion
```

Structured sections improve readability and collaboration.

---

## 🔴 Advanced Exercise 1

**Why deploying directly from notebook is risky:**

1. Notebook code is not modular
2. Hidden execution order issues
3. Debugging is difficult
4. Not scalable
5. Not production-ready

**Professional Workflow:**

1. Prototype in Jupyter
2. Move stable code to `.py` scripts
3. Write reusable functions/classes
4. Create API (FastAPI/Flask)
5. Use Docker for deployment
6. Monitor performance

Notebook = experiment
Production = structured codebase

---

## 🔴 Advanced Exercise 2

### Refactoring Strategy

**1. Clean Notebook**

* Remove unused cells
* Restart kernel and re-run
* Delete duplicated code

**2. Modularize Code**

* Move reusable functions to separate `.py` file
* Import them into notebook

Example:

```python
from preprocessing import clean_data
from model import train_model
```

**3. Move to .py Files When:**

* Project becomes large
* Multiple notebooks depend on same logic
* Preparing for deployment
* Team collaboration increases

Notebook = experimentation
Python scripts = maintainable system
