# 🛠 1.3 Data Science Tools

---

## 1️⃣ Real-World Explanation (Intuitive)

Imagine you are a **mechanic**.

You don’t fix every car using one tool.

* Wrench → Tighten bolts
* Scanner → Diagnose issue
* Oil pump → Maintain engine
* Computer system → Analyze vehicle data

Similarly, a Data Scientist uses different tools for different tasks.

---

### 🔍 Data Science = Toolbox

| Task               | Tool Type                |
| ------------------ | ------------------------ |
| Data Cleaning      | Python, R                |
| Data Storage       | SQL, Databases           |
| Data Visualization | Power BI, Tableau        |
| ML Modeling        | Scikit-learn, TensorFlow |
| Big Data           | Spark, Hadoop            |
| Collaboration      | Git, GitHub              |

Each tool solves a specific problem.

No single tool does everything.

### 🚗 Ride apps (like Uber)

# ✅ Answers – 1.3 Data Science Tools

---

## 🟢 Basic Exercise 1

**Tools to use:**

* **Extract data:** SQL
* **Analyze data:** Python (Pandas, NumPy)
* **Visualize reports:** Power BI or Tableau

**Reason:**
SQL handles databases efficiently, Python processes data, BI tools present insights to stakeholders.

---

## 🟢 Basic Exercise 2

**Why Python is more popular than R:**

1. General-purpose language (used in web, AI, automation).
2. Strong ML ecosystem (Scikit-learn, TensorFlow, PyTorch).
3. Better production and deployment support.

---

## 🟡 Intermediate Exercise 1

**Problem:**
Pandas loads data into memory → 50M rows may cause memory crash or slow performance.

**Solution:**
Use Apache Spark

**Why:**

* Distributed computing
* Handles large-scale datasets
* Parallel processing

---

## 🟡 Intermediate Exercise 2

**Tools:**

* SQL → Extract data
* Python → Clean & model
* Power BI / Tableau → Dashboard

**Workflow:**

1. Pull data using SQL
2. Process in Python
3. Export cleaned dataset
4. Build interactive dashboard
5. Share with manager

Goal: Convert code → visual insights.

---

## 🔴 Advanced Exercise 1

### Fraud Detection Tech Stack

**1. Data Storage:**
PostgreSQL / Cloud Data Warehouse

**2. Processing:**
Apache Spark (large transaction handling)

**3. Modeling:**
Python + Scikit-learn / XGBoost

**4. Deployment:**
REST API (FastAPI) + Cloud (AWS/Azure)

**5. Monitoring:**
Performance metrics dashboard + drift detection

**Justification:**
Scalable, reliable, handles imbalance, production-ready.

---

## 🔴 Advanced Exercise 2

### Real-Time Recommendation Architecture

**1. Real-Time Ingestion:**
Kafka (streaming ingestion)

**2. Processing:**
Apache Spark Streaming

**3. Model Serving:**
Python API (FastAPI)
Pre-trained collaborative filtering model

**4. Scalability:**
Cloud infrastructure with auto-scaling

**Design Logic:**

* Stream user behavior instantly
* Update recommendation features
* Serve predictions in milliseconds
* Continuously retrain offline
