# 📊 1.1 INTRODUCTION TO DATA SCIENCE

---

## 1️⃣ Real-World Explanation (Intuitive Understanding)

Imagine you own a **food delivery business in Karachi**.

Every day you collect:

* Orders
* Delivery times
* Customer ratings
* Locations
* Payment methods
* Cancellation reasons

Now imagine this:

You just *look* at the numbers.
They mean nothing.

But when you start asking:

* Why are orders low on Tuesday?
* Why do customers cancel more in one area?
* Why are delivery times longer at night?
* Which food gives best ratings?

That process — turning raw data into useful decisions — is **Data Science**.

---

### 🔍 Simple Analogy

Think of **data as raw wheat**.

* Wheat alone → not useful.
* Grind → flour.
* Cook → bread.
* Serve → food with value.

Similarly:

* Raw data → clean → analyze → model → decision.
* That full pipeline = **Data Science**.

---

### 🔹 What Data Science Actually Is

Data Science is the combination of:

| Component                  | Why It Matters             |
| -------------------------- | -------------------------- |
| Mathematics & Statistics   | To understand patterns     |
| Programming (Python, etc.) | To handle large data       |
| Domain Knowledge           | To ask the right questions |
| Communication              | To explain insights        |

It is not just coding.
It is not just machine learning.
It is not just statistics.

It is **solving real problems using data**.

---

## 2️⃣ Practical Application (How Industry Uses It)

In real companies:

### 🏦 Banks

Detect fraud transactions using transaction patterns.

### 🛍 E-commerce (like Amazon)

Recommend products based on your previous behavior.

### 🚗 Ride apps (like Uber)

Predict ride demand and adjust pricing dynamically.

### 🎬 Streaming platforms (like Netflix)

Suggest movies using recommendation systems.

### 🏥 Hospitals

Predict which patients are high-risk.

---

### What an Industry Data Scientist Actually Does

1. Understand the business problem.
2. Collect data.
3. Clean messy data.
4. Explore patterns.
5. Build models.
6. Evaluate models.
7. Deploy solution.
8. Communicate results.

Notice:
Machine learning is just **one part** of Data Science.

# ✅ Answers – 1.1 Introduction to Data Science

---

## 🟢 Basic Exercise 1

**5 Data Types to Collect:**

1. Daily sales volume (units sold)
2. Customer foot traffic
3. Product pricing changes
4. Competitor pricing/promotions

---

## 🟢 Basic Exercise 2

**Raw Data:**
Individual patient records (age, BP, test results, diagnosis).

**Information:**
Average BP of patients this month.

**Insight:**
High BP patients above age 50 are increasing → possible lifestyle issue → need awareness campaign.

---

## 🟡 Intermediate Exercise 1

**Step-by-step Data Science Process:**

1. Define problem: High cancellation in Area X.
2. Collect data: pickup time, driver distance, wait time, pricing, weather.
3. Clean data (remove errors/missing values).
4. Explore patterns (Is wait time higher there?).
5. Identify root cause.
6. Build prediction model for cancellation likelihood.
7. Suggest solution (more drivers, adjust pricing).
8. Monitor results.

---

## 🟡 Intermediate Exercise 2

**Identify At-Risk Customers:**

1. Define churn criteria (e.g., no purchase for 90 days).
2. Analyze patterns of customers who stopped buying.
3. Use features:

   * Low purchase frequency
   * Low spending
   * Long time since last purchase
4. Build churn prediction model.
5. Target high-risk customers with offers.

---

## 🔴 Advanced Exercise 1

**Major Challenge:**
Class imbalance (0.5% fraud).

**Approach:**

1. Use resampling (SMOTE / under-sampling).
2. Avoid accuracy metric (misleading).
3. Use precision, recall, F1-score, ROC-AUC.
4. Focus on minimizing false negatives (fraud missed).
5. Consider business cost of misclassification.

---

## 🔴 Advanced Exercise 2

**Movie Recommendation System**

**Data Needed:**

* User watch history
* Ratings
* Movie genres
* Watch time
* Search history

**Model Type:**

* Collaborative Filtering
* Content-Based Filtering
* Hybrid Model

**Evaluation:**

* Precision@K
* Recall@K
* CTR (Click Through Rate)

**Improve Over Time:**

* Retrain with new data
* A/B testing
* User feedback loop
